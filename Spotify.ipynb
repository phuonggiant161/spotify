# Mapping genre artist
Lấy  genre của artist qua hashtag từ hai kênh Spotipy và LastFM (ưu tiên LastFM vì kênh này đầy đủ và matching chuẩn hơn)

## Spotipy

```python
import time
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Đăng nhập API
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="****", client_secret="****"))

# Dictionary để lưu kết quả đã truy vấn (cache)
artist_cache = {}

def get_artist_genres_spotipy (artist_name):
    # Nếu nghệ sĩ đã có trong cache, trả về ngay kết quả
    if artist_name in artist_cache:
        return artist_cache[artist_name]

    # Thêm delay để tránh vượt quá rate limit (0.5 giây giữa các yêu cầu)
    time.sleep(0.5)

    try:
        # Tìm kiếm nghệ sĩ trên Spotify
        results = sp.search(q="artist:" + artist_name, type="artist", limit=1)
        items = results['artists']['items']
        genres = items[0]['genres'] if items else []
    except Exception as e:
        print(f"Lỗi khi truy vấn {artist_name}: {e}")
        genres = []

    # Lưu kết quả vào cache
    artist_cache[artist_name] = genres
    return genres
```

```python
import pandas as pd

# Lấy danh sách nghệ sĩ duy nhất từ cột 'artistName'
unique_artists_all = pd.concat([music_played['artistName'],mylibrary['artist']]).to_frame('artistName')
unique_artists = unique_artists_all['artistName'].drop_duplicates().reset_index(drop=True)


# Tạo DataFrame map tên nghệ sĩ và genre
artist_genres_df = pd.DataFrame(unique_artists, columns=['artistName'])
artist_genres_df['spotipy_genres'] = artist_genres_df['artistName'].apply(get_artist_genres_spotipy)

# Merge kết quả vào DataFrame gốc
music_played = music_played.merge(artist_genres_df, on='artistName', how='left')

```

## Last FM

```python
import requests
import time
from urllib.parse import quote

LASTFM_API_KEY = "****"  

def get_artist_genres_fm(artist_name):
    # URL encode tên nghệ sĩ để đảm bảo URL hợp lệ
    encoded_artist = quote(artist_name)
    url = f"http://ws.audioscrobbler.com/2.0/?method=artist.getInfo&artist={encoded_artist}&api_key={LASTFM_API_KEY}&format=json"

    try:
        response = requests.get(url)
        # Kiểm tra nếu API trả về HTTP status không phải 200, có thể in ra lỗi
        if response.status_code != 200:
            print(f"HTTP error for {artist_name}: {response.status_code}")
            return []
        data = response.json()
        if "artist" in data and "tags" in data["artist"]:
            tags = data["artist"]["tags"].get("tag", [])
            genres = [tag["name"] for tag in tags]
        else:
            genres = []
    except Exception as e:
        print(f"Lỗi khi truy vấn {artist_name}: {e}")
        genres = []

    # Sleep để tránh rate limit
    time.sleep(0.3)
    return genres

#example
print(get_artist_genres_fm("#SSdaRealest"))
```

```python
#mapping
# df: artist_genre.head(10)
# Tạo DataFrame mapping tên nghệ sĩ và genre
artist_genres_df['fm_genres'] = artist_genres_df['artistName'].apply(get_artist_genres_fm)
```

# Mapping country của ca sĩ

```python
# lấy country từ api của music brain
import requests

def get_artist_country_musicbrainz(artist_name):
    url = f"https://musicbrainz.org/ws/2/artist/?query={artist_name}&fmt=json"
    response = requests.get(url)
    data = response.json()

    # Kiểm tra xem có kết quả không
    if "artists" in data and len(data["artists"]) > 0:
        country = data["artists"][0].get("area", {}).get("name", "Unknown")
        return country
    return "Unknown"
```

## Last FM

```python
import requests

LASTFM_API_KEY = "*****"  # API key của bạn

def get_track_genre_lastfm(track_name, artist_name):
    url = f"http://ws.audioscrobbler.com/2.0/?method=track.getInfo&api_key={LASTFM_API_KEY}&artist={artist_name}&track={track_name}&format=json"
    try:
        response = requests.get(url)
        data = response.json()
        if "track" in data and "toptags" in data["track"]:
            tags = data["track"]["toptags"].get("tag", [])
            return [tag["name"] for tag in tags]
        return []
    except Exception as e:
        print(f"Lỗi khi truy vấn '{track_name}' của '{artist_name}': {e}")
        return []

# Áp dụng cho DataFrame
song_list["genres"] = song_list.apply(lambda row: get_track_genre_lastfm(row["track"], row["artist"]), axis=1)

# Kiểm tra kết quả
print(song_list.head())
```

```python
#đổi tên cột
song_list.rename(columns={
    "artist": "artistName",
    "track": "trackName",
}, inplace=True)
```

```python
#import csv của genre artist vào, nếu trường hợp genre song blank thì sẽ lấy từ genre artist
artist_info = pd.read_csv(r"/content/artist_info.csv")

#merge
song_list = song_list.merge(artist_info, on="artistName", how="left")

# lọc lấy final genre, nếu genre của song mà blank thì lấy genre của artist
song_list["genre_final"] = song_list.apply(lambda row: row["fm_genres"] if not row["genres"] else row["genres"], axis=1)

#xóa cột không cần thiết
track_info = song_list.drop(columns =["spotipy_genres","fm_genres","genres","country"])
```

# Tách hashtag thành nhiều dòng

```python
import ast

def convert_to_list(x):
    if isinstance(x, str):  # Nếu x là string, chuyển thành list
        try:
            return ast.literal_eval(x)  # Chuyển đổi từ chuỗi thành list thực sự
        except:
            return []  # Nếu lỗi, trả về list rỗng
    elif isinstance(x, list):  # Nếu đã là list, giữ nguyên
        return x
    else:
        return []  # Nếu là kiểu dữ liệu khác (None, NaN), trả về list rỗng

# Áp dụng hàm chuyển đổi
track_info["genre_final"] = track_info["genre_final"].apply(convert_to_list)

# Kiểm tra lại kiểu dữ liệu
print(track_info["genre_final"].apply(type).unique())  # Kết quả mong muốn: <class 'list'>
```

```python
#tách dòng
song_list_exploded = track_info.explode("genre_final").reset_index(drop=True)
```
