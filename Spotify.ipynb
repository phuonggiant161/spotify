# Mapping genre artist
L·∫•y  genre c·ªßa artist qua hashtag t·ª´ hai k√™nh Spotipy v√† LastFM (∆∞u ti√™n LastFM v√¨ k√™nh n√†y ƒë·∫ßy ƒë·ªß v√† matching chu·∫©n h∆°n)

## Spotipy

```python
import time
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# ƒêƒÉng nh·∫≠p API
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="****", client_secret="****"))

# Dictionary ƒë·ªÉ l∆∞u k·∫øt qu·∫£ ƒë√£ truy v·∫•n (cache)
artist_cache = {}

def get_artist_genres_spotipy (artist_name):
    # N·∫øu ngh·ªá sƒ© ƒë√£ c√≥ trong cache, tr·∫£ v·ªÅ ngay k·∫øt qu·∫£
    if artist_name in artist_cache:
        return artist_cache[artist_name]

    # Th√™m delay ƒë·ªÉ tr√°nh v∆∞·ª£t qu√° rate limit (0.5 gi√¢y gi·ªØa c√°c y√™u c·∫ßu)
    time.sleep(0.5)

    try:
        # T√¨m ki·∫øm ngh·ªá sƒ© tr√™n Spotify
        results = sp.search(q="artist:" + artist_name, type="artist", limit=1)
        items = results['artists']['items']
        genres = items[0]['genres'] if items else []
    except Exception as e:
        print(f"L·ªói khi truy v·∫•n {artist_name}: {e}")
        genres = []

    # L∆∞u k·∫øt qu·∫£ v√†o cache
    artist_cache[artist_name] = genres
    return genres
```

```python
import pandas as pd

# L·∫•y danh s√°ch ngh·ªá sƒ© duy nh·∫•t t·ª´ c·ªôt 'artistName'
unique_artists_all = pd.concat([music_played['artistName'],mylibrary['artist']]).to_frame('artistName')
unique_artists = unique_artists_all['artistName'].drop_duplicates().reset_index(drop=True)


# T·∫°o DataFrame map t√™n ngh·ªá sƒ© v√† genre
artist_genres_df = pd.DataFrame(unique_artists, columns=['artistName'])
artist_genres_df['spotipy_genres'] = artist_genres_df['artistName'].apply(get_artist_genres_spotipy)

# Merge k·∫øt qu·∫£ v√†o DataFrame g·ªëc
music_played = music_played.merge(artist_genres_df, on='artistName', how='left')

```

## Last FM

```python
import requests
import time
from urllib.parse import quote

LASTFM_API_KEY = "****"  

def get_artist_genres_fm(artist_name):
    # URL encode t√™n ngh·ªá sƒ© ƒë·ªÉ ƒë·∫£m b·∫£o URL h·ª£p l·ªá
    encoded_artist = quote(artist_name)
    url = f"http://ws.audioscrobbler.com/2.0/?method=artist.getInfo&artist={encoded_artist}&api_key={LASTFM_API_KEY}&format=json"

    try:
        response = requests.get(url)
        # Ki·ªÉm tra n·∫øu API tr·∫£ v·ªÅ HTTP status kh√¥ng ph·∫£i 200, c√≥ th·ªÉ in ra l·ªói
        if response.status_code != 200:
            print(f"HTTP error for {artist_name}: {response.status_code}")
            return []
        data = response.json()
        if "artist" in data and "tags" in data["artist"]:
            tags = data["artist"]["tags"].get("tag", [])
            genres = [tag["name"] for tag in tags]
        else:
            genres = []
    except Exception as e:
        print(f"L·ªói khi truy v·∫•n {artist_name}: {e}")
        genres = []

    # Sleep ƒë·ªÉ tr√°nh rate limit
    time.sleep(0.3)
    return genres

#example
print(get_artist_genres_fm("#SSdaRealest"))
```

```python
#mapping
# df: artist_genre.head(10)
# T·∫°o DataFrame mapping t√™n ngh·ªá sƒ© v√† genre
artist_genres_df['fm_genres'] = artist_genres_df['artistName'].apply(get_artist_genres_fm)
```

```python
# l·∫•y country t·ª´ api c·ªßa music brain
import requests

def get_artist_country_musicbrainz(artist_name):
    url = f"https://musicbrainz.org/ws/2/artist/?query={artist_name}&fmt=json"
    response = requests.get(url)
    data = response.json()

    # Ki·ªÉm tra xem c√≥ k·∫øt qu·∫£ kh√¥ng
    if "artists" in data and len(data["artists"]) > 0:
        country = data["artists"][0].get("area", {}).get("name", "Unknown")
        return country
    return "Unknown"
```

```python
import requests

LASTFM_API_KEY = "*****"  # API key c·ªßa b·∫°n

def get_track_genre_lastfm(track_name, artist_name):
    url = f"http://ws.audioscrobbler.com/2.0/?method=track.getInfo&api_key={LASTFM_API_KEY}&artist={artist_name}&track={track_name}&format=json"
    try:
        response = requests.get(url)
        data = response.json()
        if "track" in data and "toptags" in data["track"]:
            tags = data["track"]["toptags"].get("tag", [])
            return [tag["name"] for tag in tags]
        return []
    except Exception as e:
        print(f"L·ªói khi truy v·∫•n '{track_name}' c·ªßa '{artist_name}': {e}")
        return []

# √Åp d·ª•ng cho DataFrame
song_list["genres"] = song_list.apply(lambda row: get_track_genre_lastfm(row["track"], row["artist"]), axis=1)

# Ki·ªÉm tra k·∫øt qu·∫£
print(song_list.head())
```

# T√°ch hashtag th√†nh nhi·ªÅu d√≤ng

```python
import ast

def convert_to_list(x):
    if isinstance(x, str):  # N·∫øu x l√† string, chuy·ªÉn th√†nh list
        try:
            return ast.literal_eval(x)  # Chuy·ªÉn ƒë·ªïi t·ª´ chu·ªói th√†nh list th·ª±c s·ª±
        except:
            return []  # N·∫øu l·ªói, tr·∫£ v·ªÅ list r·ªóng
    elif isinstance(x, list):  # N·∫øu ƒë√£ l√† list, gi·ªØ nguy√™n
        return x
    else:
        return []  # N·∫øu l√† ki·ªÉu d·ªØ li·ªáu kh√°c (None, NaN), tr·∫£ v·ªÅ list r·ªóng

# √Åp d·ª•ng h√†m chuy·ªÉn ƒë·ªïi
track_info["genre_final"] = track_info["genre_final"].apply(convert_to_list)

# Ki·ªÉm tra l·∫°i ki·ªÉu d·ªØ li·ªáu
print(track_info["genre_final"].apply(type).unique())  # K·∫øt qu·∫£ mong mu·ªën: <class 'list'>
```
